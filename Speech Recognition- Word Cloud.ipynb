```python
import speech_recognition as sp
import glob2 as gb
```

```python
r= sp.Recognizer()
```

```python
#Collecting all the text from audio files
file_list=[]
n=0
for file in gb.glob('/Users/sameerrathi/Documents/MyProjects/Speech Recognition/DataSet 1/AudioData/JK/*.wav'):
    n=n+1
    try:
        with sp.AudioFile(file) as source:
            audio= r.listen(source)
            text=r.recognize_google(audio)
            file_list.append(text)
            print("reading audio file ", n)
    except Exception as e:
        print("cannot understand audio no ",n, " because of", e)
```

```python
#Cheking if the list length amount to the number of audio samples
len(file_list)
        
```

```python
#conducting an Exploratory Analysis using WordCloud
from wordcloud import WordCloud
```

```python
#preparing the data. adding a ". " after each sentence and combining all the list elememnts
Temp=[]
for text in file_list:
    temp=". "
    temp= text+temp
    Temp.append(temp)
Full_Text=[''.join(Temp)]
print(Full_Text)
```

```python
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=200, contour_width=10, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(Full_Text[0])
# Visualize the word cloud
wordcloud.to_image()
```
